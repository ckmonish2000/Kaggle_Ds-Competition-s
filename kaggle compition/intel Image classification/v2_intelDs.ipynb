```python
def convBlock(in_channel,out_channel,pool=False):
    layers=[nn.Conv2d(in_channel,out_channel,padding=1,kernel_size=3),
           nn.BatchNorm2d(out_channel),
           nn.ReLU(inplace=True)]
    if pool:layers.append(nn.MaxPool2d(2,2))
    return nn.Sequential(*layers)
```

```python
class resnet(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1=convBlock(3,64,pool=True)
        self.conv2=convBlock(64,128,pool=True)
        self.res1=nn.Sequential(convBlock(128,128,pool=False),convBlock(128,128,pool=False))
        self.conv3=convBlock(128,256,pool=True)
        self.conv4=convBlock(256,300,pool=True)
        self.res2=nn.Sequential(convBlock(300,300,pool=False),convBlock(300,300,pool=False))
        self.classify=nn.Sequential(nn.MaxPool2d(2,2),nn.Flatten(),nn.Linear(300*7*7,6))
    def forward(self,x):
        out=self.conv1(x)
        out=self.conv2(out)
        out=self.res1(out)+out
        out=self.conv3(out)
        out=self.conv4(out)
        out=self.res2(out)+out
        out=self.classify(out)
        return out
```

```python
def accuracy(output,labels):
    _,preds=torch.max(output,dim=1)
    return torch.tensor(torch.sum(preds==labels).item()/len(output))
```

```python
def fit(epoch,max_lr,wd,GC,model,loss,dl,train=True):
    torch.cuda.empty_cache()
    if(train==True):
        model.train()
        optim=torch.optim.SGD(model.parameters(),weight_decay=wd,lr=0.1)
        sch=torch.optim.lr_scheduler.OneCycleLR(optim,max_lr=max_lr,epochs=epoch,steps_per_epoch=len(dl))
#         sch=torch.optim.lr_scheduler.StepLR(optim,step_size=7,gamma=0.1)
        for x in range(epoch):
            for i,j in dl:
                i=i.to(device)
                j=j.to(device)
                pred=model(i)
                ls=loss(pred,j)
                ls.backward()
                
                if GC:
                    nn.utils.clip_grad_value_(model.parameters(),GC)
                optim.step()
                optim.zero_grad()
                sch.step()
                acc=accuracy(pred,j)
            print(f"epoch={x} loss={ls} acc={acc}")
    else:
        model.eval()
        for x in range(epoch):
            for i,j in dl:
                i=i.to(device)
                j=j.to(device)
                pred=model(i)
                ls=loss(pred,j)
                acc=accuracy(pred,j)
        print(f"epoch={x} loss={ls} acc={acc}")
        
            
                
                
            
```
